```python
#Create Dataframe from file
urls = pd.read_csv('url (1).csv')
urls.head()
```

```python
#Count number of episodes
number = urls.count()
number = number['url']
number
```

```python
#Char to search for
char = ['Leonard: ', 'Sheldon: ', 'Penny: ', 'Howard: ', 'Raj: ','Leslie: ',
       'Bernadette: ', 'Amy: ', 'Stuart: ', 'Emily: ', 'Wil: ', 'Mary: ','Zack: ',
       'Bert: ', 'Janine: ']
```

```python
# Searches for char in urls
i = 0


number = urls.count()
number = number['url']


while i < number:
    
    req = urllib2.Request(urls.loc[i,"url"]) #reads the html for each url
    response = urllib2.urlopen(req)
    html = response.read()
    
    j=0
    while j < len(char):
        charvalue = html.count(char[j]) #searches for char in url html
        urls.loc[i,char[j]] = charvalue
        j = j+1
        
    print i
    i += 1
print 'complete'
    
```
