```python
#Create Dataframe from file
urls = pd.read_csv('AD (5).csv')
```

```python
#Count number of episodes
number = urls.count()
number = number['url']
number
```

```python
#Char to search for
char = ['Narrator:', 'Michael:','G.O.B.:', 'George Michael:', 'Tobias:', 'Lindsay:','Buster:',
       'Maeby:', 'Sr.:', 'Oscar:', 'Lucille:','John Beard:',
       'Stan Sitwell:', 'Starla:', 'Barry:', 'Lucille 2:', 'Ann:', 'Steve Holt',
       'Rebel:',  'Kitty:', 'Lupe:', 'Annyong:', 'Mort Meyers:']
```

```python
# Searches for char in urls
i = 0


number = urls.count()
number = number['url']


while i < number:
    
    req = urllib2.Request(urls.loc[i,"url"]) #reads the html for each url
    response = urllib2.urlopen(req)
    html = response.read()
    
    j=0
    while j < len(char):
        charvalue = html.count(char[j]) #searches for char in url html
        urls.loc[i,char[j]] = charvalue
        j = j+1
        
    print i
    i += 1
print 'complete'
    
```
