```python
#Create Dataframe from file
urls = pd.read_csv('scrubs (5).csv')
urls.head()
```

```python
#Count number of episodes
number = urls.count()
number = number['Title']
number
```

```python
#Char to search 
char = ['Turk:', 'Dr. Cox:', 'Dr. Kelso:', 'J.D.:', 'Elliot:', 'Janitor:','Carla:', 'Todd:', 'Ted:', 'Nurse Roberts:', 
        'Jordan:', 'Doug:', 'Keith:', 'Dr. Wen:', 'Dr. Mickhead:', 'Lloyd:', 'Lonnie:', 'Crispin:', 'Randall:', 'Roy:', 
        'Leonard:', 'Snoop Dogg Intern:', 'Sean:', 'Danni:', 'Gloria:', 'Lisa:']
```

```python
# Searches for char in urls
i = 0


number = urls.count()
number = number['Title']

    
while i < number:
    
    req = urllib2.Request(urls.loc[i,"url"]) #reads the html for each url
    response = urllib2.urlopen(req)
    html = response.read()
    
    j=0
    while j < len(char):
        charvalue = html.count(char[j]) #searches for char in url html
        urls.loc[i,char[j]] = charvalue
        j = j+1
        
    print i
    i += 1
print 'complete'
    
```
